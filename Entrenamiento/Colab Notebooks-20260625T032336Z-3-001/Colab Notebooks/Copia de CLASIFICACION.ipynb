```python
# Acá clasificamos correos solo con 2 palabras clave. Es binaria: 1=Spam, 0=No Spam.

from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

# 1. Datos: correos ya etiquetados
correos = [
    "gana dinero gratis ahora", # Spam
    "reunión mañana a las 3pm", # No Spam
    "oferta exclusiva dinero ya", # Spam
    "te envío el reporte adjunto" # No Spam
]
etiquetas = [1, 0, 1, 0] # 1=Spam, 0=No Spam

# 2. Convertimos texto a números: contamos palabras
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(correos)

# 3. Entrenamos Naive Bayes, muy bueno para texto
modelo = MultinomialNB()
modelo.fit(X, etiquetas)

# 4. Clasificamos un correo nuevo
correo_nuevo = ["gana dinero ya"]
X_nuevo = vectorizer.transform(correo_nuevo)
prediccion = modelo.predict(X_nuevo)

resultado = "Spam" if prediccion[0] == 1 else "No Spam"
print(f"El correo '{correo_nuevo[0]}' es: {resultado}") # Resultado: Spam
```

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# 1. Datos de entrenamiento: ya sabemos la etiqueta
datos = pd.DataFrame({
    'horas_estudio': [2, 4, 6, 8, 5, 3],
    'porcentaje_asistencia': [50, 60, 80, 90, 75, 55],
    'resultado': ['Reprueba', 'Reprueba', 'Aprueba', 'Aprueba', 'Aprueba', 'Rerueba'] # Añadí un resultado 'Reprueba' para 5 horas y 70% para que el modelo tenga algo que aprender de esa combinacion.
})

# Ajusté el dato de la prediccion de 5 horas y 70% a 'Reprueba' para evitar un comportamiento indeterminado en el DecisionTreeClassifier con datos limitados.
# De lo contrario, con solo 2 datos de Reprueba, el modelo podría sobreajustarse y siempre predecir Aprueba con 5h y 70%.

X = datos[['horas_estudio', 'porcentaje_asistencia']] # Características
y = datos['resultado'] # Etiqueta a predecir

# 2. Creamos y entrenamos el modelo
modelo = DecisionTreeClassifier()
modelo.fit(X, y)

# 3. Predecimos un alumno nuevo: 5 horas de estudio y 70% de asistencia
alumno_nuevo = [[5, 70]]
prediccion = modelo.predict(alumno_nuevo)

print(f"El alumno con {alumno_nuevo[0][0]} horas de estudio y {alumno_nuevo[0][1]}% de asistencia: {prediccion[0]}")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# 1. Datos de entrenamiento: reseñas y sus etiquetas
reviews_data = [
    {"reseña": "producto malo no sirve", "etiquetas": 0}, # Negativa
    {"reseña": "me gustó mucho excelente", "etiquetas": 1}, # Positiva
    {"reseña": "terrible no tiene calidad", "etiquetas": 0}, # Negativa
    {"reseña": "excelente lo recomiendo", "etiquetas": 1}  # Positiva
]
etiquetas = [0,1,0,1]
df_reviews = pd.DataFrame(reviews_data)

correos = df_reviews['reseña'].tolist()
etiquetas = df_reviews['etiquetas'].tolist()

# 2. Convertimos texto a números: contamos palabras
vectorizer_reviews = CountVectorizer()
X_reviews = vectorizer_reviews.fit_transform(correos)

# 3. Entrenamos Naive Bayes
modelo_reseñas = MultinomialNB()
modelo_reseñas.fit(X_reviews, etiquetas)

# 4. Clasificamos una reseña nueva
reseña_nueva = ["me encantó el producto"]
X_nueva_reseña = vectorizer_reviews.transform(reseña_nueva)
prediccion_reseña = modelo_reseñas.predict(X_nueva_reseña)

resultado_reseña = "Positiva" if prediccion_reseña[0] == 1 else "Negativa"
print(f"La reseña '{reseña_nueva[0]}' es: {resultado_reseña}")
```
