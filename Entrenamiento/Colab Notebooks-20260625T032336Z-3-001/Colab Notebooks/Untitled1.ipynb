```python
import matplotlib.pyplot as plt
import numpy as np

# Simulación de puntos de búsqueda (donde se encuentra a la mascota)
# Mayor concentración cerca del origen (0,0)
np.random.seed(42)
n_puntos = 100
# Probabilidad de encontrar a la mascota disminuye con la distancia
distancia = np.random.exponential(scale=300, size=n_puntos) # metros
angulo = np.random.uniform(0, 2*np.pi, size=n_puntos)

x = distancia * np.cos(angulo)
y = distancia * np.sin(angulo)

plt.figure(figsize=(8,8))
plt.scatter(x, y, s=distancia/10 + 5, c=distancia, cmap='autumn_r', alpha=0.6)
plt.scatter(0, 0, c='blue', s=200, marker='*', label='Punto de Escape')
plt.title('Probabilidad de localización de mascota extraviada')
plt.xlabel('Distancia (metros)')
plt.ylabel('Distancia (metros)')
plt.colorbar(label='Metros desde el escape')
plt.legend()
plt.grid(True)
plt.show()
```
