```python
import gradio as gr

def calculator(num1, operation, num2):
    if operation == '+':
        return num1 + num2
    elif operation == '-':
        return num1 - num2
    elif operation == '*':
        return num1 * num2
    elif operation == '/':
        if num2 == 0:
            return "Error: División por cero"
        return num1 / num2

calculator_demo = gr.Interface(
    fn=calculator,
    inputs=[
        gr.Number(label="Primer Número"),
        gr.Dropdown(["+", "-", "*", "/"], label="Operación"),
        gr.Number(label="Segundo Número")
    ],
    outputs="text",
    title="Calculadora Sencilla",
    description="Introduce dos números y selecciona una operación."
)

calculator_demo.launch()
```
